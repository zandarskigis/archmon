## NDVI/SAVI Forecasting with Recurrent Neural Networks (RNN)

This notebook uses precomputed NDVI and SAVI rasters to forecast vegetation change patterns over time using a recurrent neural network (RNN) model. The goal is to detect areas where predicted vegetation indices diverge from actual observations, potentially indicating disturbances related to environmental or human activity (e.g., low-level flight training).

Unlike earlier versions, this workflow assumes that NDVI and SAVI have already been generated and exported as GeoTIFFs. This allows us to focus entirely on the time-series modeling process.

### Workflow Overview:
- Load NDVI (and optionally SAVI) rasters organized by year and month
- Stack NDVI into a temporal array suitable for RNN input
- Split data into:
  - **Training Set**: 2017–2021
  - **Validation Set**: 2022
  - **Test Set**: 2023–2024
- Train an RNN model to forecast NDVI into future years (2025–2027)
- Compare 2025 predictions against observed imagery to evaluate forecasting accuracy
- Identify areas of significant deviation to guide archaeological or environmental monitoring

```python
import glob
import rasterio

# Load preprocessed NDVI rasters
ndvi_files = sorted(glob.glob(os.path.join(ndvi_input_dir, "NDVI_*.tif")))

ndvi_stack = []
ndvi_dates = []

for f in ndvi_files:
    fname = os.path.basename(f).lower()
    
    # Skip any files from 2025 or with 'clipped' in the name
    if "2025" in fname or "clipped" in fname:
        continue

    with rasterio.open(f) as src:
        arr = src.read(1)
        if arr.shape != (109, 136):
            print(f"Skipping {f}: unexpected shape {arr.shape}")
            continue
        ndvi_stack.append(arr)
        ndvi_dates.append(os.path.basename(f).replace("NDVI_", "").replace(".tif", ""))

ndvi_stack = np.stack(ndvi_stack)  # shape = (time, rows, cols)
```

```python
#from calendar import month_name
from datetime import datetime
from calendar import month_name

# Month lookup dictionary
month_lookup = {name.lower(): f"{i:02d}" for i, name in enumerate(month_name) if name}

# Initialize
ndvi_arrays = []
timestamps = []

for f in ndvi_files:
    filename = os.path.basename(f).replace("NDVI_", "").replace(".tif", "")
    try:
        year, month_text = filename.split("-")
        month_num = month_lookup.get(month_text.lower())
        if month_num is None:
            raise ValueError("Unrecognized month")

        date_obj = datetime.strptime(f"{year}-{month_num}", "%Y-%m")
        with rasterio.open(f) as src:
            arr = src.read(1)
            if arr.shape != (109, 136):
                print(f"Skipping {f}: unexpected shape {arr.shape}")
                continue
            ndvi_arrays.append(arr)
            timestamps.append(date_obj)

    except ValueError:
        print(f"Skipping file with non-parseable date: {filename}")

# Stack into a 3D array
ndvi_stack = np.stack(ndvi_arrays, axis=0)  # shape: (time, height, width)

# Map year indices
year_index = [t.year for t in timestamps]
train_idx = [i for i, y in enumerate(year_index) if 2017 <= y <= 2021]
val_idx   = [i for i, y in enumerate(year_index) if y == 2022]
test_idx  = [i for i, y in enumerate(year_index) if y in (2023, 2024)]

# Split the stack
ndvi_train = ndvi_stack[train_idx]
ndvi_val   = ndvi_stack[val_idx]
ndvi_test  = ndvi_stack[test_idx]

# Confirm shapes
print("NDVI Stack Shape:", ndvi_stack.shape)
print("Train/Val/Test Lengths:", len(train_idx), len(val_idx), len(test_idx))
```

## RNN Forecasting Objective

The goal of this modeling workflow is to train a recurrent neural network (RNN) on NDVI and/or SAVI time series data from stable, pre-flight years (2017–2024), and use it to forecast vegetation conditions for 2025, 2026, and 2027.

These predictions will be compared against actual Sentinel-2 imagery (as it becomes available) to:

- **Detect deviations or anomalies** that emerge following the onset of low-level flight training (beginning in 2025)
- **Identify areas of potential flight-induced vegetation stress or erosion**
- **Guide field inspections** — for example, comparing predicted vs. observed NDVI for September 2025 prior to scheduled October site visits

This evaluation framework is designed to highlight where observed vegetation change differs from model expectations based on historical stability. Such deviations may indicate indirect impacts from training activity or broader site instability.

## Sequence Generation for RNN

We will generate sequences for both NDVI and SAVI independently. Each index can be modeled separately or compared side-by-side, allowing us to evaluate their individual and combined predictive performance.

### Sequence Types:
- **12 → 1**: Use 12 months of input to predict the next 1 month
- **24 → 3**: Use 24 months of input to predict the next 3 months

This design allows short-term and medium-range forecasting while testing different input horizons.

### Output Format:
Sequences will be saved as `.npy` files with consistent naming patterns, such as:
- `ndvi_train_X_12_1.npy`, `ndvi_train_Y_12_1.npy`
- `savi_train_X_24_3.npy`, `savi_train_Y_24_3.npy`

### Workflow Steps:
- Create sequences from `ndvi_stack` or `savi_stack`
- Normalize values (e.g., to range [0, 1] or [-1, 1])
- Flatten spatial dimensions (or optionally use patches for spatial modeling)
- Save resulting arrays to the RNN input folder for later model training

This modular approach provides flexibility for experimenting with different input-output time windows and evaluation strategies.

```python
import os
import numpy as np
import joblib
from sklearn.preprocessing import MinMaxScaler

# Define output directory for sequences
out_dir = os.path.join("..", "data", "5HF737", "rnn_inputs")
os.makedirs(out_dir, exist_ok=True)

# --- Helper: Normalize and flatten NDVI stack ---
def normalize_and_flatten(stack):
    n_time, height, width = stack.shape
    reshaped = stack.reshape(n_time, -1)  # (time, pixels)
    scaler = MinMaxScaler()
    norm = scaler.fit_transform(reshaped)
    norm_reshaped = norm.reshape(n_time, height, width)
    return norm_reshaped, scaler

# --- Normalize NDVI splits using training scaler ---
ndvi_train_norm, ndvi_scaler = normalize_and_flatten(ndvi_train)
ndvi_val_norm, _ = normalize_and_flatten(ndvi_val)
ndvi_test_norm, _ = normalize_and_flatten(ndvi_test)

# Save the scaler for later inverse transforms
joblib.dump(ndvi_scaler, os.path.join(out_dir, "ndvi_minmax_scaler.pkl"))
print("Saved NDVI scaler to ndvi_minmax_scaler.pkl")

# --- Sequence generation ---
def create_sequences(stack, input_len, pred_len):
    X, Y = [], []
    for i in range(len(stack) - input_len - pred_len + 1):
        x_seq = stack[i:i+input_len]
        y_seq = stack[i+input_len:i+input_len+pred_len]
        X.append(x_seq.reshape(input_len, -1))  # Flatten spatial dims
        Y.append(y_seq.reshape(pred_len, -1))
    return np.array(X), np.array(Y)

# --- Generate and save sequences ---
seqs = [(12, 1), (24, 3)]  # Define input → output windows

for input_len, pred_len in seqs:
    for split_name, split_data in zip(['train', 'val', 'test'],
                                      [ndvi_train_norm, ndvi_val_norm, ndvi_test_norm]):
        X, Y = create_sequences(split_data, input_len, pred_len)

        x_path = os.path.join(out_dir, f"ndvi_{split_name}_X_{input_len}_{pred_len}.npy")
        y_path = os.path.join(out_dir, f"ndvi_{split_name}_Y_{input_len}_{pred_len}.npy")
        
        np.save(x_path, X)
        np.save(y_path, Y)

        print(f"Saved: {os.path.basename(x_path)} (shape {X.shape})")
        print(f"Saved: {os.path.basename(y_path)} (shape {Y.shape})")
```

```python
import glob

# Reuse the month_lookup from earlier
# month_lookup = {name.lower(): f"{i:02d}" for i, name in enumerate(month_name) if name}

# List all SAVI files in the folder
savi_files = sorted(glob.glob(os.path.join(savi_input_dir, "SAVI_*.tif")))

savi_arrays = []
savi_timestamps = []

for f in savi_files:
    filename = os.path.basename(f).replace("SAVI_", "").replace(".tif", "")
    
    # Skip 2025 or any misformatted names
    if "2025" in filename or "clipped" in filename.lower():
        continue

    try:
        year, month_text = filename.split("-")
        month_num = month_lookup.get(month_text.lower())
        if month_num is None:
            raise ValueError("Unrecognized month")

        date_obj = datetime.strptime(f"{year}-{month_num}", "%Y-%m")
        with rasterio.open(f) as src:
            arr = src.read(1)
            if arr.shape != (109, 136):
                print(f"Skipping {f}: unexpected shape {arr.shape}")
                continue
            savi_arrays.append(arr)
            savi_timestamps.append(date_obj)

    except ValueError:
        print(f"Skipping SAVI file with non-parseable date: {filename}")

# Stack into 3D array
savi_stack = np.stack(savi_arrays, axis=0)  # shape: (time, height, width)

# Split by year
savi_year_index = [t.year for t in savi_timestamps]
savi_train_idx = [i for i, y in enumerate(savi_year_index) if 2017 <= y <= 2021]
savi_val_idx   = [i for i, y in enumerate(savi_year_index) if y == 2022]
savi_test_idx  = [i for i, y in enumerate(savi_year_index) if y in (2023, 2024)]

savi_train = savi_stack[savi_train_idx]
savi_val   = savi_stack[savi_val_idx]
savi_test  = savi_stack[savi_test_idx]

# Print result summary
print("SAVI Stack Shape:", savi_stack.shape)
print("Train/Val/Test Lengths:", len(savi_train_idx), len(savi_val_idx), len(savi_test_idx))
```

```python
# Reuse same output directory
out_dir = os.path.join("..", "data", "5HF737", "rnn_inputs")
os.makedirs(out_dir, exist_ok=True)

# Normalize SAVI splits using training scaler
savi_train_norm, savi_scaler = normalize_and_flatten(savi_train)
savi_val_norm, _ = normalize_and_flatten(savi_val)
savi_test_norm, _ = normalize_and_flatten(savi_test)

# Save the SAVI scaler for inverse transforms later
joblib.dump(savi_scaler, os.path.join(out_dir, "savi_minmax_scaler.pkl"))
print("Saved SAVI scaler to savi_minmax_scaler.pkl")

# Create and save sequences for each split
for input_len, pred_len in seqs:
    for split_name, split_data in zip(['train', 'val', 'test'],
                                      [savi_train_norm, savi_val_norm, savi_test_norm]):
        X, Y = create_sequences(split_data, input_len, pred_len)

        x_path = os.path.join(out_dir, f"savi_{split_name}_X_{input_len}_{pred_len}.npy")
        y_path = os.path.join(out_dir, f"savi_{split_name}_Y_{input_len}_{pred_len}.npy")

        np.save(x_path, X)
        np.save(y_path, Y)

        print(f"Saved: {os.path.basename(x_path)} (shape {X.shape})")
        print(f"Saved: {os.path.basename(y_path)} (shape {Y.shape})")
```
