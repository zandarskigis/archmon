# NDVI 12→1 LSTM Model Pipeline

This notebook trains a Long Short-Term Memory (LSTM) neural network on NDVI sequences, using 12 months of input to predict NDVI for the next month (12→1 setup).

It assumes that NDVI sequences were preprocessed and saved as `.npy` arrays in a prior notebook.

### Project Directory Structure

Assuming `archmon/` is the repository root:

- NDVI/SAVI sequence files:  
  `archmon/data/5HF737/rnn_inputs/`  
  (e.g., `ndvi_train_X_12_1.npy`, `ndvi_test_Y_12_1.npy`, etc.)

- NDVI rasters for evaluation:  
  `archmon/data/5HF737/ndvi_by_month/`

- SAVI rasters (optional):  
  `archmon/data/5HF737/savi_by_month/`

This setup ensures compatibility when running locally or sharing via GitHub. The model will produce spatial NDVI predictions and compare them with observed NDVI imagery from 2025 onward to detect anomalies potentially associated with low-level flight activity.


```python
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# --- Load NDVI Sequence Data ---
X_train = np.load("../data/5HF737/rnn_inputs/ndvi_train_X_12_1.npy")  # (samples, 12, 14824)
Y_train = np.load("../data/5HF737/rnn_inputs/ndvi_train_Y_12_1.npy")  # (samples, 1, 14824)
X_val   = np.load("../data/5HF737/rnn_inputs/ndvi_val_X_12_1.npy")    # might be empty
Y_val   = np.load("../data/5HF737/rnn_inputs/ndvi_val_Y_12_1.npy")    # might be empty

# --- Clip and clean ---
X_train = np.nan_to_num(np.clip(X_train, 0, 1))
Y_train = np.nan_to_num(np.clip(Y_train, 0, 1))[:, 0, :]  # remove pred_len dim

X_val = np.nan_to_num(np.clip(X_val, 0, 1))
if Y_val.size > 0:
    Y_val = np.nan_to_num(np.clip(Y_val, 0, 1))[:, 0, :]
    validation_data = (X_val, Y_val)
else:
    Y_val = None
    validation_data = None
    print("⚠️  No validation data found. Proceeding without validation.")

# --- Confirm shapes ---
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
if Y_val is not None:
    print("X_val shape:", X_val.shape)
    print("Y_val shape:", Y_val.shape)

# --- Define LSTM Model ---
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False),
    Dense(Y_train.shape[1])  # predict all pixels
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
model.summary()

# --- Train Model ---
history = model.fit(
    X_train, Y_train,
    validation_data=validation_data,
    epochs=30,
    batch_size=16,
    verbose=1
)

# --- Save Model ---
os.makedirs("../outputs/models/", exist_ok=True)
model.save("../outputs/models/ndvi_lstm_12_1.h5")
print("✅ Model saved to ../outputs/models/ndvi_lstm_12_1.h5")

# --- Plot Loss Curve ---
plt.plot(history.history['loss'], label='Train Loss')
if 'val_loss' in history.history:
    plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# NDVI 12→1 LSTM Model Pipeline

This notebook trains and evaluates a Long Short-Term Memory (LSTM) model to predict NDVI using a 12→1 sequence format:  
**12 months of NDVI input → 1 month of predicted NDVI output**.

It also generates predictions for the 2023–2024 test period and sets up for 2025 NDVI forecasting to assess potential low-level flight impacts on vegetation.

### Directory Structure (Relative to `archmon/` repo root)

- **Preprocessed NDVI/SAVI sequences**  
  `archmon/data/5HF737/rnn_inputs/`  
  (e.g., `ndvi_train_X_12_1.npy`, `ndvi_test_Y_12_1.npy`)

- **NDVI raster imagery for testing and comparison**  
  `archmon/data/5HF737/ndvi_by_month/`

- **SAVI rasters (optional or for future use)**  
  `archmon/data/5HF737/savi_by_month/`

All output predictions and models will be saved in `archmon/outputs/`.

```python
import numpy as np
import matplotlib.pyplot as plt

# === Load and clean test data ===
X_test = np.load("../data/5HF737/rnn_inputs/ndvi_test_X_12_1.npy")
Y_test = np.load("../data/5HF737/rnn_inputs/ndvi_test_Y_12_1.npy")

X_test = np.nan_to_num(np.clip(X_test, 0, 1))
Y_test = np.nan_to_num(np.clip(Y_test, 0, 1))

# If Y_test is (samples, 1, 14824), squeeze it
if Y_test.ndim == 3:
    Y_test = Y_test[:, 0, :]  # Now (samples, 14824)

print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)

# === Predict ===
predictions = model.predict(X_test)
print("Predictions shape:", predictions.shape)

# === Visualize one prediction ===
sample_idx = 0
pred_raster = predictions[sample_idx].reshape(109, 136)
actual_raster = Y_test[sample_idx].reshape(109, 136)

# Plot predicted, actual, and difference
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(pred_raster, cmap='YlGn', vmin=0, vmax=1)
axes[0].set_title("Predicted NDVI")
axes[1].imshow(actual_raster, cmap='YlGn', vmin=0, vmax=1)
axes[1].set_title("Actual NDVI")
axes[2].imshow(pred_raster - actual_raster, cmap='RdBu', vmin=-0.5, vmax=0.5)
axes[2].set_title("NDVI Difference (Pred - Actual)")

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Rebuild rasters if they’re not already defined in the notebook context
sample_idx = 0
pred_raster = predictions[sample_idx].reshape(109, 136)
actual_raster = Y_test[sample_idx].reshape(109, 136)
diff_raster = pred_raster - actual_raster

# Enhanced plot with colorbars and context
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Predicted NDVI
im0 = axes[0].imshow(pred_raster, cmap='YlGn', vmin=0, vmax=1)
axes[0].set_title("Predicted NDVI")
cbar0 = plt.colorbar(im0, ax=axes[0], shrink=0.8)
cbar0.set_label("NDVI (0–1)")
axes[0].text(0, -10, "Predicted from LSTM model based on past 12 months", fontsize=9)

# Actual NDVI
im1 = axes[1].imshow(actual_raster, cmap='YlGn', vmin=0, vmax=1)
axes[1].set_title("Actual NDVI")
cbar1 = plt.colorbar(im1, ax=axes[1], shrink=0.8)
cbar1.set_label("NDVI (0–1)")
axes[1].text(0, -10, "Observed NDVI from satellite imagery", fontsize=9)

# Difference map
im2 = axes[2].imshow(diff_raster, cmap='RdBu', vmin=-1, vmax=1)
axes[2].set_title("Difference (Pred - Actual)")
cbar2 = plt.colorbar(im2, ax=axes[2], shrink=0.8)
cbar2.set_label("NDVI Delta")
axes[2].text(0, -10, "Blues = underpredicted, Reds = overpredicted", fontsize=9)

# Clean axes
for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.savefig("ndvi_prediction_panel.png", dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# === Parameters ===
forecast_steps = 12
raster_shape = (109, 136)
pixel_count = raster_shape[0] * raster_shape[1]

# === Load last test sequence ===
X_test = np.load("../data/5HF737/rnn_inputs/ndvi_test_X_12_1.npy")
X_test = np.nan_to_num(np.clip(X_test, 0, 1))

# Use the final test sequence as seed (Dec 2024)
forecast_sequence = X_test[-1].copy()
forecast_results = []

# === Recursive Forecast ===
for i in range(forecast_steps):
    input_seq = forecast_sequence[-12:].reshape(1, 12, pixel_count)
    pred = model.predict(input_seq)[0]  # shape: (14824,)
    forecast_results.append(pred)
    forecast_sequence = np.vstack([forecast_sequence, pred[np.newaxis, :]])

# === Save forecast output ===
forecast_array = np.array(forecast_results)  # shape: (12, 14824)
forecast_rasters = forecast_array.reshape((12, *raster_shape))

# Ensure output directory exists
os.makedirs("../outputs/forecasts/", exist_ok=True)
np.save("../outputs/forecasts/ndvi_forecast_2025.npy", forecast_rasters.reshape(12, -1))
print("✅ Forecast saved to ../outputs/forecasts/ndvi_forecast_2025.npy")

# === Plot all 12 forecast months ===
fig, axes = plt.subplots(3, 4, figsize=(16, 10))
for i, ax in enumerate(axes.flat):
    im = ax.imshow(forecast_rasters[i], cmap='YlGn', vmin=0, vmax=1)
    ax.set_title(f"Forecast Month {i+1} (2025)")
    ax.axis('off')

fig.suptitle("LSTM Forecasted NDVI – 2025 (12→1)", fontsize=16)
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.8)
cbar.set_label("NDVI (0–1)")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import os

# === Load SAVI test sequence ===
X_savi_test = np.load("../data/5HF737/rnn_inputs/savi_test_X_12_1.npy")
X_savi_test = np.nan_to_num(np.clip(X_savi_test, 0, 1))

# === Parameters ===
savi_forecast_steps = 12
raster_shape = (109, 136)
pixel_count = raster_shape[0] * raster_shape[1]

# === Use last test sequence (Dec 2024) as seed ===
savi_forecast_sequence = X_savi_test[-1].copy()
savi_forecast_results = []

# === Recursive forecast ===
for i in range(savi_forecast_steps):
    input_seq = savi_forecast_sequence[-12:].reshape(1, 12, pixel_count)
    pred = model.predict(input_seq)[0]  # shape: (14824,)
    savi_forecast_results.append(pred)
    savi_forecast_sequence = np.vstack([savi_forecast_sequence, pred[np.newaxis, :]])

# === Reshape into 12 rasters (109x136) ===
savi_forecast_array = np.array(savi_forecast_results)
savi_forecast_rasters = savi_forecast_array.reshape((12, *raster_shape))

# === Save for downstream use ===
os.makedirs("../outputs/forecasts/", exist_ok=True)
np.save("../outputs/forecasts/savi_forecast_2025.npy", savi_forecast_rasters.reshape(12, -1))
print("✅ SAVI forecast saved to ../outputs/forecasts/savi_forecast_2025.npy")
```

```python
import numpy as np
import rasterio
from pathlib import Path
import matplotlib.pyplot as plt

# === CONFIG ===
predicted_ndvi = forecast_rasters  # Already in memory from LSTM forecast
actual_dir = Path("../data/5HF737/ndvi_by_month")
diff_dir = Path("../outputs/forecasts/ndvi_diffs")
diff_dir.mkdir(parents=True, exist_ok=True)

# === Difference + Metrics ===
metrics = {}

for i, month in enumerate(["january", "february", "march"]):
    actual_path = actual_dir / f"ndvi_2025_{month}_clipped.tif"
    if not actual_path.exists():
        print(f"Missing: {actual_path.name}")
        continue

    with rasterio.open(actual_path) as src:
        actual = src.read(1).astype("float32")
        profile = src.profile

    predicted = predicted_ndvi[i]
    
    if actual.shape != predicted.shape:
        print(f"Shape mismatch for {month}: {actual.shape} vs {predicted.shape}")
        continue

    diff = predicted - actual
    diff = np.clip(diff, -1, 1)

    # Save difference raster
    diff_path = diff_dir / f"ndvi_diff_2025_{month}.tif"
    profile.update(dtype=rasterio.float32, count=1)
    with rasterio.open(diff_path, 'w', **profile) as dst:
        dst.write(diff, 1)

    # Metrics
    mae = np.mean(np.abs(diff))
    rmse = np.sqrt(np.mean(diff ** 2))
    metrics[month] = {"MAE": mae, "RMSE": rmse}

    # Visualization
    plt.figure(figsize=(6, 4))
    im = plt.imshow(diff, cmap="RdBu", vmin=-0.5, vmax=0.5)
    plt.title(f"NDVI Δ (Pred - Actual): {month.capitalize()}")
    cbar = plt.colorbar(im, shrink=0.8)
    cbar.set_label("NDVI Delta")
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import rasterio
from pathlib import Path
import matplotlib.pyplot as plt

# === CONFIG ===
predicted_ndvi = forecast_rasters  # Already in memory from LSTM forecast
actual_dir = Path("../data/5HF737/ndvi_by_month")
diff_dir = Path("../outputs/forecasts/ndvi_diffs")
diff_dir.mkdir(parents=True, exist_ok=True)

# === Difference + Metrics ===
metrics = {}

for i, month in enumerate(["january", "february", "march"]):
    actual_path = actual_dir / f"ndvi_2025_{month}_clipped.tif"
    if not actual_path.exists():
        print(f"Missing: {actual_path.name}")
        continue

    with rasterio.open(actual_path) as src:
        actual = src.read(1).astype("float32")
        profile = src.profile

    predicted = predicted_ndvi[i]

    if actual.shape != predicted.shape:
        print(f"Shape mismatch for {month}: {actual.shape} vs {predicted.shape}")
        continue

    diff = predicted - actual
    diff = np.clip(diff, -1, 1)

    # Save difference raster
    diff_path = diff_dir / f"ndvi_diff_2025_{month}.tif"
    profile.update(dtype=rasterio.float32, count=1)
    with rasterio.open(diff_path, 'w', **profile) as dst:
        dst.write(diff, 1)

    # === Metrics ===
    mae = np.mean(np.abs(diff))
    rmse = np.sqrt(np.mean(diff ** 2))
    metrics[month] = {"MAE": mae, "RMSE": rmse}

    # === Enhanced Visualization ===
    fig, ax = plt.subplots(figsize=(8, 5.5))
    im = ax.imshow(diff, cmap="RdBu", vmin=-0.5, vmax=0.5)
    ax.set_title(f"NDVI Prediction Δ (Predicted - Actual): {month.capitalize()}", fontsize=13)

    # Colorbar
    cbar = plt.colorbar(im, ax=ax, shrink=0.8)
    cbar.set_label("NDVI Δ Value", fontsize=10)
    cbar.ax.tick_params(labelsize=9)

    # Metrics
    metric_str = f"MAE: {mae:.3f} | RMSE: {rmse:.3f}"
    ax.text(0.02, 0.97, metric_str, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

    # Interpretation note
    ax.text(0.02, 0.88,
            "Blue = Underprediction\nRed = Overprediction",
            transform=ax.transAxes, fontsize=9, color='gray',
            verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

    ax.axis("off")
    plt.tight_layout()

    # Save annotated PNG
    output_img = diff_dir / f"ndvi_diff_plot_{month}.png"
    plt.savefig(output_img, dpi=300, bbox_inches='tight', transparent=True)
    plt.show()
```

## Conclusion

This notebook demonstrates the successful application of LSTM-based forecasting to detect landscape-scale vegetation changes at archaeological sites. By comparing predicted NDVI values for 2025 with observed Sentinel-2 imagery, we identified spatial patterns of deviation that may correspond to environmental disturbance or anthropogenic impact, such as low-level military flights.

Despite the strength of the model, the visual and quantitative differences between forecasted and actual NDVI speak for themselves. RMSE and MAE values provide additional context, but the spatial maps clearly reveal where vegetation trends diverged from expected baselines. These difference maps can now be used to guide field inspection, prioritize monitoring zones, or trigger more advanced machine learning analyses.

While this version omits slope masking for simplicity, future iterations will reintroduce topographic controls and explore ensemble modeling with SAVI and other indices. This pipeline is designed to be modular, reproducible, and scalable across additional time periods and sites.

This work underscores the power of temporal modeling for heritage monitoring — not just detecting change, but anticipating it.
